```python
import pandas as pd
```

# Fetch content of all sections generated by READMEClassifier

```python
# these files where generated by the readmeclassifier into a light sql tables. We exported them into csv
readme_section_content = '../../data/ghtorrent_readme_classifications/generated_data/target_section_content.csv'
readme_section_overview = '../../data/ghtorrent_readme_classifications/generated_data/target_section_overview.csv'
```

```python
overview_df = pd.read_csv(readme_section_overview, encoding="utf-8") 
print(len(overview_df.file_id.unique()))
overview_df.set_index(['file_id', 'section_id'], inplace=True)
```

```python
content_df = pd.read_csv(readme_section_content, encoding="utf-8")  
content_df.set_index(['file_id', 'section_id'], inplace = True)
```

```python
content_df
```

```python
content_overview_df = overview_df.join(content_df, how='inner')
```

```python
content_overview_df.reset_index(inplace=True )
content_overview_df.set_index('local_readme_file', inplace=True )
len(content_overview_df.file_id.unique())
```

```python
content_overview_df.head()
```

# Save Content of  WhatWhy Sections

## get repos in use

```python
whatwhy_repos = '../../data/processed/repositories_with-readme_what-why-code.csv'
whatwhy_repos_df = pd.read_csv(whatwhy_repos, encoding="utf-8") 
```

```python
len(whatwhy_repos_df)
```

```python
def set_md_fullname(row):
    md_name = '{}.md'.format(row['readme_file_name']) 
    row['local_readme_file'] = md_name
    return row
```

```python
whatwhy_repos_df = whatwhy_repos_df.apply(set_md_fullname, axis=1)
```

```python
whatwhy_repos_df.set_index('local_readme_file', inplace = True)
```

## get content df data for what why

```python
content_overview_df['section_code'].fillna('', inplace=True)
whatwhy_content_df = content_overview_df[content_overview_df['section_code'].str.contains('1')]
```

```python
len(whatwhy_content_df.file_id.unique())
```

```python
len(whatwhy_content_df)
```

## join all data

```python
combined_df = whatwhy_content_df.join(whatwhy_repos_df, how='inner',  lsuffix='_readme')
```
